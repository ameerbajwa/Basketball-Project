```python
import csv
import numpy as np
import pandas as pd

from bs4 import BeautifulSoup, Comment
from fake_useragent import UserAgent
import time
import requests, re
import random
import string
import pickle
```

```python
# List of NBA teams 

teams = ['ATL', 'BOS', 'BRK', 'NJN', 'CHO', 'CHH', 'CHA', 'CHI', 'CLE', 'DAL', 'DEN', 'DET', 'GSW', 'HOU', 'IND', 
         'LAC', 'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOH', 'NOK', 'NOP', 'NYK', 'OKC', 'SEA', 'ORL', 'PHI', 'PHO', 
         'POR', 'SAC', 'SAS', 'TOR', 'UTA', 'WAS', 'WSB']

years = list(range(1990,2019))
```

### Creating soup objects from each page of basketball-reference.com

```python
# Function to scrape urls of NBA teams

us = UserAgent()
user_agent = {'User-Agent':us.random}

def get_team_info(team, year):
    url = 'https://www.basketball-reference.com/teams/'+team+'/'+str(year)+'.html'
    response = requests.get(url, headers=user_agent)
    page = response.text
    print (url)
    time.sleep(3)
    return page
```

```python
team_info = {}

for team in teams:
    for year in years:
        soup = BeautifulSoup(get_team_info(team, year), 'lxml')
        team_info[team+'_'+str(year)] = soup
```

###  Grabbing the indices for the dataframe 

```python
p_url = 'https://www.basketball-reference.com/teams/BRK/2018.html'
p_response = requests.get(p_url, headers=user_agent)
p_page = p_response.text
p_soup = BeautifulSoup(p_page, 'lxml')
```

```python
comments = p_soup.find_all(text=lambda text: isinstance(text, Comment))

rx_team_stat = re.compile(r'<table.+?id="team_and_opponent".+?>[\s\S]+?</table>')
rx_team_misc = re.compile(r'<table.+?id="team_misc".+?[\s\S]+?</table>')
```

```python
for comment in comments:
    try:
        team_stat_table = rx_team_stat.search(comment.string).group(0)
        stat_table = BeautifulSoup(team_stat_table, 'lxml')
        stats_info = [p for p in stat_table.find_all('tr')]
    except:
        pass
    try:
        team_misc_table = rx_team_misc.search(comment.string).group(0)
        misc_table = BeautifulSoup(team_misc_table, 'lxml')
        misc_info = [m for m in misc_table.find_all('tr')]
    except:
        pass
```

```python
indices_misc = [a.text for a in misc_info[1].find_all('th')]
indices = [a.text for a in stats_info[0].find_all('th')]
```

```python
team_indices = indices + indices_misc
```

### Extracting Information from each soup object

```python
# Grabbing all the commented tables from the soup objects for the team stats

team_year_stats = {}

for team_year, info in team_info.items():
    team_year_stats[team_year] = []
    
    comments = info.find_all(text=lambda text:isinstance(text, Comment))

    rx_team_stat = re.compile(r'<table.+?id="team_and_opponent".+?>[\s\S]+?</table>')
    rx_team_misc = re.compile(r'<table.+?id="team_misc".+?[\s\S]+?</table>')
    
    for comment in comments:
        try:
            team_stat_table = rx_team_stat.search(comment.string).group(0)
            stat_table = BeautifulSoup(team_stat_table, 'lxml')
            stats_info = [p for p in stat_table.find_all('tr')]
            
            for i in stats_info[1:]:
                team_stats = i.find_all('td')
                stats = [s.text for s in team_stats]
    
                team_year_stats[team_year].extend(stats)
        except:
            pass
        try:
            team_misc_table = rx_team_misc.search(comment.string).group(0)
            misc_table = BeautifulSoup(team_misc_table, 'lxml')
            misc_info = [m for m in misc_table.find_all('tr')]
            
            for i in misc_info[2:]:
                misc_stats = i.find_all('td')
                m_stats = [s.text for s in misc_stats]
    
                team_year_stats[team_year].extend(m_stats)
        except:
            pass
```
